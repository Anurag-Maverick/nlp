```python
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import TweetTokenizer
```

# Preprocessing Tweet: removing Repeated Tweet, hyperlinks, hashtags

```python
tweet2 = re.sub(r'^RT[\s]+', '',tweet)
tweet2 = re.sub(r'https?:\/\/.*[\r\n]*', '', tweet2)
tweet2 = re.sub(r'#', '', tweet2)

tweet2
```

# Tokenizer

```python
tokenizer = TweetTokenizer(preserve_case=True, reduce_len=True, strip_handles=True)
tweet3 = tokenizer.tokenize(tweet2)
tweet3
```

# Listing all english stopwords

```python
english_stopwords = stopwords.words("english")
english_stopwords
```

```python
import string
string.punctuation
```

# Clean Tweet

```python
clean_tweet = []
for word in tweet3:
    if word not in english_stopwords and word not in string.punctuation:
        clean_tweet.append(word)

clean_tweet
```

```python
stemmer = PorterStemmer()
stemmed_words = []

for word in clean_tweet:
    stemmed_word = stemmer.stem(word)
    stemmed_words.append(stemmed_word)
    
stemmed_words
```

```python
def process_tweet(tweet):
    tweet = re.sub(r'^RT[\s]+', '',tweet)
    tweet = re.sub(r'https?:\/\/.*[\r\n]*', '', tweet)
    tweet = re.sub(r'#', '', tweet)
    
    english_stopwords = stopwords.words("english")
    tokenizer = TweetTokenizer(preserve_case=True, reduce_len=True, strip_handles=True)
    stemmer = PorterStemmer()
    
    tweet_tokens = tokenizer.tokenize(tweet)
    
    clean_tweet = []
    for word in tweet_tokens:
        if word not in english_stopwords and word not in string.punctuation:
            word_stem = stemmer.stem(word)
            clean_tweet.append(word_stem)
    
    return clean_tweet
```

```python
tweet = all_positive_tweets[2277]
cleaned_words = process_tweet(tweet)
cleaned_words
```
